```python
import matplotlib.pyplot as plt
import numpy as np

# .   
x = np.linspace(0, 10, 100)
y = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x, y, linestyle='-', marker='', linewidth=2)
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
plt.show()
```

```python
# .   (Scatter)
x = np.random.rand(50)
y = np.random.rand(50)
fig, ax = plt.subplots()
ax.scatter(x, y)
plt.show()
```

```python
# .   (Bar)
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 12, 36]
fig, ax = plt.subplots()
ax.bar(categories, values)
plt.show()
```

```python
# . 
data = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(data, bins=20)
plt.show()
```

```python
# .     Figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8,4))
ax1.plot(x, np.cos(x)); ax1.set_title('cos(x)')
ax2.hist(data, bins=15);   ax2.set_title('Histogram')
plt.tight_layout()
plt.show()
```
